<h5>Seaborn Bar plot</h5>
1. Write a program to draw bar plot of sex against survived for a dataset given in the url

```python
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("titanic")
sns.catplot(x = "sex", y = "survived", data = data, kind='bar')
plt.show()
```

<h5>Seaborn Scatterplot</h5>
1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a
url

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
#data =  pd.read_csv('tips.csv')
data = sns.load_dataset('tips')
sns.scatterplot(x='day', y='total_bill', data=data, hue='smoker')
plt.show()
```

<h5>Seaborn BoxPlot</h5>
1. Write a program to draw box plot of life expectancy by continent for a data set given in a FiveYearData csv file

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("FiveYearData.csv")
data.boxplot(by ='continent', column =['lifeExp'])
plt.show()
```

2. Write a program to draw a box plot of day by tips for a dataset given in a tips.csv file

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("tips.csv")
data.boxplot(by ='day', column =['tip'])
plt.show()
```

<h5>Seaborn Swarm Plot</h5>
1. Write a program to draw a swarm plot of total bill against size for a given dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
data = sns.load_dataset("tips")
#sns.swarmplot(x="total_bill", y="size", data=data)
sns.stripplot(x="total_bill", y="size", data=data)
plt.show()
```

2. Write a program to draw swarm plot of “total bill” against day for a dataset given in tips.csv file

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
data = sns.load_dataset("tips")
sns.swarmplot(x="total_bill", y="day", data=data)
plt.show()
```
