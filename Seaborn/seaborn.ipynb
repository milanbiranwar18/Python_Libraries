<h5> Seaborn Bar plot</h5>
1. Write a program to draw bar plot of sex against survived for a dataset given in the url

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
#sns.set_style('darkgrid')
#data = pd.read_csv('titanic.csv')
data = sns.load_dataset("titanic")
#sns.barplot(x = "sex", y = "survived", data = data)
sns.catplot(x = "sex", y = "survived", data = data, kind = 'bar')
plt.show()
```

Seaborn Scatterplot
1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a tips.csv

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset('tips')
sns.scatterplot(x='day', y='total_bill', data=data, hue='smoker')
plt.show()
```

<h5>Seaborn Violin Plot</h5>
1. Write a program to draw a violin plot of sex against total_bill for a given dataset


```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset('tips')
sns.violinplot(x='sex', y='total_bill', data=data, hue='smoker')
plt.show()
```

Seaborn BoxPlot
1. Write a program to draw box plot of life expectancy by continent for a data set given in a FiveYearData.csv

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("FiveYearData.csv")
df.boxplot(by ='continent', column =['lifeExp'])
plt.show()
```

2. Write a program to draw a box plot of day by tips for a dataset given in a tips.csv

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("tips.csv")
df.boxplot(by ='day', column =['tip'], grid = False)
plt.show()
```

<h5>Seaborn Swarm Plot</h5>
1. Write a program to draw a swarm plot of total bill against size for a given dataset


```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
# sns.swarmplot(x="total_bill", y="size", data=tips)
sns.stripplot(x="total_bill", y="size", data=tips)
plt.show()
```

2. Write a program to draw swarm plot of “total bill” against day for a dataset given in tips.csv

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
sns.swarmplot(x="total_bill", y="day", data=data)
plt.show()
```
